```python
import pandas as pd 
import re
from PyPDF2 import PdfReader 
```

```python
rutaArch = "/Users/brandonvasm/Desktop/Analisis Anual 2018 ETAS.pdf"
```

```python
documento = PdfReader(rutaArch)
```

```python
pagina = documento.pages[4]
texto = pagina.extract_text() 
texto 
```

```python
patron = r'\n(\d+)\s+(\d+)\s+([\s\S]+?)\s{2,}([\s\S]+?)\s{2,}([\s\S]+?)\s+(\d+)'
```

```python
datos = re.findall(patron, texto)
datos 
```

```python
etas = pd.DataFrame(datos, columns=['No', 'SE', 'Area de Salud', 'ETA reportada', 'Fuente de contagio', 'No. de casos'])
etas
```
